```python
import torch
import gym
import matplotlib.pyplot as plt
import time
from gym.envs.registration import register
"""
ACTIONS
LEFT = 0
DOWN = 1
RIGHT = 2
UP = 3

STATES
4 x 4 = 16

[0,15]
"""
plt.style.use('ggplot')
env = gym.make('FrozenLake-v0', is_slippery=True)

gamma = 0.95
learning_rate = 0.9


num_episodes = 1000
steps_total = []
rewards_total = []
Q = torch.zeros(number_of_states, number_of_actions)

number_of_states = env.observation_space.n
number_of_actions = env.action_space.n

for episode in range(num_episodes):
    # State
    state = env.reset()
    
    step = 0

    #for step in range(100):
    while True:
        step +=1 
        
        # Random Action
        random_values = Q[state] + torch.rand(1, number_of_actions) /1000
        
        action = int(torch.max(random_values, 1)[1][0])
        
        
        new_state, reward, done, info = env.step(action)
        
        # Bellmann Equation
        Q[state, action] = (1-learning_rate) * Q[state, action] + learning_rate * (reward + gamma * torch.max(Q[new_state]))
                
        state = new_state
        
        #time.sleep(0)
        #env.render()
        
        if done:
            steps_total.append(step)
            rewards_total.append(reward)
            #print('Episode finished after %i steps' % step)
            break
#env.close()
print(Q)

print("Percentage of episodes finished succesfully: {} %".format(sum(rewards_total)*100/num_episodes))
print("Percentage of episodes finished succesfully (Last 100 episodes): {} %".format(sum(rewards_total[-100:]*100)/100))


#print(steps_total)
print("Average number of steps: %.2f" % (sum(steps_total) / num_episodes))
print("Average number of steps(Last 100 steps): %.2f" % (sum(steps_total[-100:])/100))

plt.figure(figsize=(12,5))
plt.title("Rewards")
plt.bar(torch.arange(len(rewards_total)), rewards_total, alpha=0.6, color='green')
plt.show()

plt.figure(figsize=(12,5))
plt.title("Steps / Episode length ")
plt.bar(torch.arange(len(steps_total)), steps_total, alpha=0.6, color='red')
plt.show()
```

# Adaptive E-Greedy

```python
import torch
import gym
import matplotlib.pyplot as plt
import time
from gym.envs.registration import register
"""
ACTIONS
LEFT = 0
DOWN = 1
RIGHT = 2
UP = 3

STATES
4 x 4 = 16

[0,15]
"""
plt.style.use('ggplot')
env = gym.make('FrozenLake-v0', is_slippery=False)

gamma = 0.90
#learning_rate = 0.9

egreedy = 0.7
egreedy_final = 0.1
egreedy_decay = 0.999
egreedy_total = []


num_episodes = 1000
steps_total = []
rewards_total = []
Q = torch.zeros(number_of_states, number_of_actions)

number_of_states = env.observation_space.n
number_of_actions = env.action_space.n

for episode in range(num_episodes):
    # State
    state = env.reset()
    
    step = 0

    #for step in range(100):
    while True:
        step +=1  
        
        random_for_greedy = torch.rand(1)[0]
        
        # E-Greedy
        if random_for_greedy > egreedy :
            random_values = Q[state] + torch.rand(1, number_of_actions) /1000
            action = int(torch.max(random_values, 1)[1][0])
        else:
            action = env.action_space.sample()
        
        if egreedy > egreedy_final:
            egreedy *= egreedy_decay
        
        new_state, reward, done, info = env.step(action)
        
        # Bellmann Equation
        Q[state, action] = reward + gamma * torch.max(Q[new_state])
                
        state = new_state
        
        #time.sleep(0)
        #env.render()
        
        if done:
            steps_total.append(step)
            rewards_total.append(reward)
            egreedy_total.append(egreedy)
            #print('Episode finished after %i steps' % step)
            break
#env.close()
print(Q)

print("Percentage of episodes finished succesfully: {} %".format(sum(rewards_total)*100/num_episodes))
print("Percentage of episodes finished succesfully (Last 100 episodes): {} %".format(sum(rewards_total[-100:]*100)/100))


#print(steps_total)
print("Average number of steps: %.2f" % (sum(steps_total) / num_episodes))
print("Average number of steps(Last 100 steps): %.2f" % (sum(steps_total[-100:])/100))

plt.figure(figsize=(12,5))
plt.title("Rewards")
plt.bar(torch.arange(len(rewards_total)), rewards_total, alpha=0.6, color='green')
plt.show()

plt.figure(figsize=(12,5))
plt.title("Steps / Episode length")
plt.bar(torch.arange(len(steps_total)), steps_total, alpha=0.6, color='red')
plt.show()

plt.figure(figsize=(12,5))
plt.title("E-Greedy Values ")
plt.bar(torch.arange(len(egreedy_total)), egreedy_total, alpha=0.6, color='blue')
plt.show()
```
